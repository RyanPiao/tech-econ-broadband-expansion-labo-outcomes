# Step 3 EDA Notebook

This notebook documents exploratory diagnostics for the county-year panel built on Step 2.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..').resolve()
df = pd.read_csv(ROOT / 'data_analysis' / 'county_year_panel.csv')
df.head()
```

```python
key_vars = ['remote_work_share','broadband_sub_share','digital_emp_share','median_hh_income','unemployment_rate','population']
df[key_vars].describe().T
```

```python
yearly = df.groupby('year', as_index=False).agg(
    remote_work_mean=('remote_work_share','mean'),
    broadband_mean=('broadband_sub_share','mean'),
    digital_emp_mean=('digital_emp_share','mean')
)
yearly
```

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(yearly['year'], yearly['remote_work_mean'], marker='o', label='Remote work')
ax.plot(yearly['year'], yearly['broadband_mean'], marker='o', label='Broadband')
ax.plot(yearly['year'], yearly['digital_emp_mean'], marker='o', label='Digital employment')
ax.legend(); ax.set_xlabel('Year'); ax.set_ylabel('Share'); ax.set_title('Step3 EDA Trends')
plt.show()
```
